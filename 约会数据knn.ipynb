```python
# 文本矩阵化
filename = "machinelearning/Ch02/datingTestSet.txt"
from numpy import *
import operator
def to_matrix(filename):
    love_tag = {'largeDoses':3,'smallDoses':2,'didntLike':1}
    file = open(filename)
    readlines = file.readlines()
    shape_of_matrix = len(readlines)
    matrix = zeros((shape_of_matrix,3))
    labels = []
    index = 0
    for readline in readlines:
        readline = readline.strip()
        readline = readline.split('\t')
        matrix[index,:] = readline[0:3]
        if (readline[-1].isdigit()):
            labels.append(int(readline[-1]))
        else:
            labels.append(love_tag.get(readline[-1]))
        index += 1
    return matrix,labels
```

```python
# 矩阵归一化
def matrixnorm(matrix):
    maxs = matrix.max(0)
    mins = matrix.min(0)
    ranges = maxs - mins 
    #公式为 (matrix-min)/(max-min)
    shape = matrix.shape[0]
    first_step = matrix - tile(mins,(shape,1))
    normmatrix = first_step/tile(ranges,(shape,1))
    return normmatrix,ranges,mins
```

```python
# knn分类算法
def knn(matrix,input_vector,labels,k):
    shape = matrix.shape[0]
    vector = tile(input_vector,(shape,1))
    d = (matrix - vector)**2
    dd = (d.sum(axis = 1))**0.5
    sorted_dd_index = dd.argsort()
    vote_labels = {}
    for m in range(k):
        vote = labels[sorted_dd_index[m]]
        vote_labels[vote] = vote_labels.get(vote,0)+1
    vote_labels = sorted(vote_labels.items(),key = operator.itemgetter(1),reverse=True)
    return vote_labels[0][0]
```

```python
# 约会数据测试算法
filename = "machinelearning/Ch02/datingTestSet.txt"
def testclassification(filename):
    matrix,labels = to_matrix(filename)
    matrix= matrixnorm(matrix)
    ratio = 0.5
    matrix_shape = matrix.shape[0]
    test_shape = int(matrix_shape*ratio)
    error_vote = 0.0
    for z in range(test_shape):
        testlabels = knn(matrix[test_shape:matrix_shape,:],matrix[z,:],labels[test_shape:matrix_shape],3)
        reallabels = labels[z]
        print("测试所得的标签为%d，实际标签为%d"%(testlabels,reallabels))
        if testlabels != reallabels:
            error_vote +=1
    error_ratio = error_vote/test_shape
    print("模型的错误率为%f"%(error_ratio))
    print("模型共有%d个样本，%d样本出错。"%(test_shape,error_vote))
```

```python
testclassification(filename)
```

```python
def youlikeornot(filename):
    result = ["你很喜欢","一般喜欢","有点喜欢"]
    a = float(input("你一般花百分之多少的时间用来打游戏？"))
    b = float(input("一年的飞行时间大概有多少？"))
    c= float(input("一年消耗冰淇淋多少公升？"))
    vector = array([a,b,c])
    matrix,labels = to_matrix(filename)
    normmatrix,ranges,mins = matrixnorm(matrix)
    vector = (vector - mins)/ranges
    result_index = knn(matrix,vector,labels,3)
    likeornot = result[result_index-1]
    print("你可能{}这个人".format(likeornot))
    
```
