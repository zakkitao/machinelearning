```python
# 将32*32 转化为1024
from numpy import *
filename = 'machinelearning/Ch02/testDigits/0_13.txt'
def to_vector(filename):
    vector = zeros((1,1024))
    file = open(filename)
    for i in range(32):
        each_line = file.readline()
        for j in range(32):
            vector[0,i*32+j] = int(each_line[j])
    return vector
a = to_vector(filename)     
```

```python
def knn(matrix,input_vector,labels,k):
    shape = matrix.shape[0]
    vector = tile(input_vector,(shape,1))
    d = (matrix - vector)**2
    dd = (d.sum(axis = 1))**0.5
    sorted_dd_index = dd.argsort()
    vote_labels = {}
    for m in range(k):
        vote = labels[sorted_dd_index[m]]
        vote_labels[vote] = vote_labels.get(vote,0)+1
    vote_labels = sorted(vote_labels.items(),key = operator.itemgetter(1),reverse=True)
    return vote_labels[0][0]
```

```python
from os import listdir
import operator

def shouxieshibie():
    labels = []
    filelist = listdir("machinelearning/Ch02/trainingDigits")
    m = len(filelist)
    trainingmat = zeros((m,1024))
    for i in range(m):
        filenamestr = filelist[i]
        filestr = filenamestr.split('.')[0]
        classnumstr = int(filestr.split('_')[0])
        labels.append(classnumstr)
        trainingmat[i,:] = to_vector('machinelearning/Ch02/trainingDigits/%s'%filenamestr)
    testfilelist = listdir('machinelearning/Ch02/testDigits')
    errorcount = 0.0
    mtest = len(testfilelist)
    for i in range(mtest):
        filenamestr = testfilelist[i]
        filestr = filenamestr.split('.')[0]
        classnumstr = int(filestr.split('_')[0])
        vectorundertest = to_vector('machinelearning/Ch02/testDigits/%s'%filenamestr)
        testlabels = knn(trainingmat,vectorundertest,labels,3)
        reallabels = classnumstr
        print("测试所得的标签为%d，实际标签为%d"%(testlabels,reallabels))
        if testlabels != reallabels:
            errorcount +=1
    error_ratio = errorcount/mtest
    print("模型的错误率为%f"%(error_ratio))
    print("模型共有%d个样本，%d样本出错。"%(mtest,errorcount))
```

```python
shouxieshibie()
```
